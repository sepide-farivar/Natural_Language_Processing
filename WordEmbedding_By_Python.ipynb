```python
sentences = ['This YouTube channel is perfect',
            'Your videos are quite good',
            'I enjoy this channel',
            'I did not learn much from your videos',
            'Excellent',
            'poor work',
            'it is not a good channel']
```

```python
vector_size = 500
```

```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
embedded_docs = [one_hot(sent,vector_size) for sent in sentences]
```

```python
embedded_docs
```

```python
max_ped = 8
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
encode_docs=pad_sequences(embedded_docs, padding='pre', maxlen=max_ped)
encode_docs
```

```python
from tensorflow.keras.layers import Embedding
```

```python
from tensorflow.keras.models import Sequential
```

```python
dim =8
```

```python
model= Sequential()
```

```python
model.add(Embedding(vector_size, dim, input_length=max_ped))
```

```python
model.compile(optimizer= 'adam', loss='mse')
```

```python
model.summary()
```

```python
embedded_docs
```

```python
model.predict(encode_docs)
```
